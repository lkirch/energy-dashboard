```python
# change detection 
# Convert data into (time, change-value, duration) so that an appliance's power and on-duration can be derived.

tsShifted = tsElec.shift(1, freq='min')
#print tsElec[1:]
#print tsShifted[0:-1]

tsFrame = DataFrame({'tsElec':tsElec[1:].values, 'tsShifted':tsShifted[0:-1].values}, index=tsElec.index[1:])
tsFrame['hour'] = tsFrame.index.hour
tsFrame['day'] = tsFrame.index.weekday
tsFrame['month'] = tsFrame.index.month
tsFrame['temp'] = tsWeather.resample('1min', fill_method='pad')

#changeBand = 1.2 # 20% change counts as a change
#tsFrame['same'] =  ((tsFrame['tsElec']/tsFrame['tsShifted'] > (1/changeBand)) & (tsFrame['tsElec']/tsFrame['tsShifted'] < changeBand)) 
tsFrame['same'] = abs(tsFrame['tsElec'] - tsFrame['tsShifted']) < 0.1
tsFrame['cvalue'] = (~tsFrame['same'])*(tsFrame['tsElec'] - tsFrame['tsShifted']) # change recorded or considered zero
#tsFrame['cvalue'] = (~tsFrame['same'])*(tsFrame['tsElec']) # new value
#print tsFrame



# ramp and duation detector, duration in minutes
# How long the given change value persists, within the change band.
def duration_det(mylist):
    ramptime = 3 # allow 3 minutes of ramp time for any appliance
    outlist = []
    zerolen = 1
    pasti = 0
    pastval = 0
    runningval = 0
    for i in range(len(mylist)):
        curval = mylist[i]
        poscur = curval > 0
        pospast = pastval > 0
        negcur = curval < 0
        negpast = pastval < 0        
        if (abs(curval) < 0.01):  #instead of comparing with zero
            zerolen += 1
        elif ((zerolen <= ramptime) and ((poscur and pospast) or (negcur and negpast))): # combine
            pastval = pastval + curval
            zerolen += 1
        else: # current value is non-zero and current value has different sign or ramptime exceeded
            outlist.append((pastval, zerolen))
            if (zerolen > 1):
                for j in range(pasti + 1, i):
                    outlist.append((0, 0))
            pasti = i
            zerolen = 1
            pastval = curval
    # end of the list, so fill it  up
    outlist.append((pastval, zerolen))
    if (zerolen > 1):
        for j in range(pasti + 1, i+1):  # i+ 1 here because last value of i needs to be entered
            outlist.append((0,0))
    return outlist 


outlist = duration_det(tsFrame['cvalue'].values)
tsFrame['change'] = [i[0] for i in outlist]
tsFrame['duration'] = [i[1] for i in outlist]


#Quantization
quant_00 = 0.1 # values less than this are quantized to 0
quant_01 = 1.0 # value less than this are quantized to +-0.1
quant_05 = 5.0 # values less than this are quantized to +-0.25
# all higher values are quantized to +-0.5

def quantizer(x):
    if (abs(x) < quant_00):
        y = 0
    elif (abs(x) < quant_01):
        y = round(x,1)
    elif (abs(x) < quant_05): 
        y = round(x*4)/4  
    else:
        y = round(x*2)/2
    return y

# change values quantized 
cvalues = tsFrame['change']
cvaluesQuantized = cvalues.map(lambda x: quantizer(x))
tsFrame['cvalueq'] = cvaluesQuantized

def reconstruct(change):
    value = 0
    outlist = []
    for i in range(len(change)):
        if (change[i] != 0):
            value = value + change[i]
        outlist.append(value)
    return outlist

tsFrame['tsElecNew'] = tsFrame['tsElec'].map(lambda x: quantizer(x))
# print a histogram to see the quantization process

cvaluesqnonzero = tsFrame['cvalueq'][tsFrame['cvalueq'] > 0]
#print cvaluesqnonzero.describe()
fig0, ax0 = plt.subplots(1,1, figsize=(20,5))
cvaluesqnonzero.hist(bins=100, ax=ax0)

#visualize a time segment with time-series plots
# findings: 
# adjust for 2-3 minute ramp time. Will miss out pulses due to 0.4W limit. Need to get change right.
# if +ve change == -ve change then duration is accurate. Match each -ve change with +ve change
# create a column for #of changes every 15 minutes and avg of +ve change - indicates high frequency changes
# know digital filters and signal processing



```
