```python
import os
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline

os.chdir("C:/Users/nitpatwa/Google Drive/data science/w210/project/pecan")
#print os.getcwd()
```

```python
dfElec = pd.read_table("2013-elec-id-6673.csv", sep=';')
dfElec = dfElec.fillna(0)
dfElec['localminute']= dfElec['localminute'].map(lambda x: x[0:19])
#print dfElec['localminute'][0]

#print small['use'][0]
use = dfElec['use']
#print use.describe()

localdatetime = dfElec['localminute']
datevalues = pd.to_datetime(localdatetime.values)
#print type(datevalues)
#print type(pd.to_datetime(localdatetime.values)) converted data into datetime series
# print use.values # convert series into array. series is a dictionary so values work

tsElec = Series(use.values, index=datevalues)
#print type(tsElec), tsElec.index
```
